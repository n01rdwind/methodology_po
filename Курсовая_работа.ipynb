тест Шапиро-Уилка

```python
import numpy as np
from scipy.stats import chi2_contingency, shapiro, norm
import matplotlib.pyplot as plt
import math as mt

# Вычисляем количество интервалов (можете изменить в соответствии с вашими предпочтениями)
num_bins = round(1 + 3.322 * mt.log10(len(set(data))))

# Строим гистограмму
hist, bin_edges = np.histogram(data, bins=num_bins)

# Вычисляем ожидаемые частоты для нормального распределения
expected_freq = len(data) / num_bins

# Проводим тест хи-квадрат
chi2_stat, p_value, _, _ = chi2_contingency([hist, [expected_freq] * num_bins])

# Выводим результаты теста
print("Хи-квадрат статистика:", chi2_stat)
print("p-значение:", p_value)

# Строим график плотности нормального распределения
plt.hist(data, bins=num_bins, edgecolor='black', alpha=0.7, density=True, label='Гистограмма данных')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='Плотность нормального распределения')

plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Интервальный ряд распределения с плотностью нормального распределения')
plt.legend()
plt.show()

# Проводим тест Шапиро-Уилка
stat, p_value = shapiro(data)

# Выводим результаты теста
print("Статистика теста:", stat)
print("p-значение:", p_value)

# Определяем уровень значимости
alpha = 0.05

# Проверяем нулевую гипотезу
if p_value > alpha:
    print("Не отвергаем нулевую гипотезу: данные распределены нормально")
else:
    print("Отвергаем нулевую гипотезу: данные не распределены нормально")
```
